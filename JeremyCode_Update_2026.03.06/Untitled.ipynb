```python
import numpy as np

predictions = np.load("predictions_JeremyCode_RevisedSplit.npy", allow_pickle=True).item()
```

```python
from sklearn.metrics import roc_auc_score

test_labels = predictions["test"]["labels"]
test_probs  = predictions["test"]["probs"]

test_auc = roc_auc_score(test_labels, test_probs[:, 1])

print("Test AUC:", test_auc)
```
